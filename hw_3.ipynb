### 3. Задание (в программе)

2.эллипса,

```python
import matplotlib.pyplot as plt
import numpy as np

# График окружности радиусом R.
fig = plt.figure(figsize=(6., 6.))

an = np.linspace(0, 2 * np.pi, 100)
R = 20
plt.plot(R * np.cos(an), R * np.sin(an))

plt.show()
```

Для пострения эллипса используем matplotlib.patches

Документация:
https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.patches.Ellipse.html#matplotlib.patches.Ellipse

```python
import matplotlib.patches as mpatches

fig = plt.figure()
fig = plt.figure(figsize=(6., 6.))
ax = fig.add_subplot(1, 1, 1)

# Построим окружность с центром (0.7, 0.2) и радиусом 0.15
circ_1 = plt.Circle((0.7, 0.2), 0.15, linewidth=2, fill=False)
# Построим круг с центром (0.5, 0.5) и радиусом 0.1
circ_2 = plt.Circle((0.5, 0.5), 0.1, color='g', alpha=0.5)
# Построим эллипс
xcenter, ycenter = 0.38, 0.55
width, height = 0.2, 0.5
angle = -30
circ_3 = mpatches.Ellipse((xcenter, ycenter), width, height,
                 angle=angle, linewidth=3, fill=False, zorder=2)

ax.add_patch(circ_1)
ax.add_patch(circ_2)
ax.add_patch(circ_3)

plt.show()
```

#### График поверхности 2-го порядка в пространстве: Параболоид

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = (X**2)/2 + (Y**2)/3  # Параболоид

ax.plot_surface(X, Y, Z)

ax.scatter(0, 0, 0,'z',50,'red')
plt.title(f'Параболоид Z = (X**2)/2 + (Y**2)/3',color="blue")
show()
```

График поверхности 2-го порядка в пространстве: Гиперболический параболоид

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = (X**2)/2 - (Y**2)/3  # Параболоид

ax.plot_surface(X, Y, Z)

ax.scatter(0, 0, 0,'z',50,'red')
plt.title(f'Гиперболический параболоид Z = (X**2)/2 - (Y**2)/3',color="blue")
show()
```

```python
# перевод полярных координат в декартовы
def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return x, y

pol2cart(3, 120)
```

### 3.4. Задание (в программе)
1)Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1


2)Решите систему уравнений и неравенств:

y = x2 – 1

exp(x) + x∙(1 – y) > 1

***Решение***

1) Записываем уравнения так, чтобы в правой части был 0:

y - x2 + 1 = 0

exp(x) + x∙(1 – y) - 1 = 0

Эти выражения оформляем как функции и функции передаем в параметры численного метода.

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1, y1 =  fsolve(equations, (-2, 1))
              
print (x1, y1)
```

Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import rcParams
rcParams["figure.figsize"] = 10, 5  # Установим масштаб изображения графика с помощью библиотеки pylab

%config InlineBackend.figure_format = 'svg' # Поменяем настройки для большей четкости изображений
```

```python
x = np.linspace(-50, 50, 201)  # Зададим значения переменной x
plt.plot(x, np.cos(1*x), label="cos(1*x)")  # для первой кривой значение k=1 
plt.plot(x, np.cos(0.3*x), label="cos(0.3*x)")  # для второй кривой значение k=0.3

plt.title('График y(k,x)=cos(k∙x)')  # Добавим название графика
plt.xlabel('Переменная x')  # добавим название оси абцисс
plt.ylabel('Функция y(x)')  # добавим название оси ординат

plt.axis([-60, 60, -1.1, 1.1])  # Зададим желаемые границы по осям
plt.legend(loc="upper right")

plt.show()
```
